### Applying PCA, ISOMAP, LLE, T-SNE on MNIST data

#### Loading required packages

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt 
from sklearn import manifold
```

#### Generating the MNIST data

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
X = digits.data
y = digits.target
```

```python
y[:4]
```

```python
def plot_clustering(X_red, X, labels, title=None):
    x_min, x_max = np.min(X_red, axis=0), np.max(X_red, axis=0)
    X_red = (X_red - x_min) / (x_max - x_min)

    plt.figure(figsize=(6, 4))
    for i in range(X_red.shape[0]):
        plt.text(X_red[i, 0], X_red[i, 1], str(labels[i]),
                 color=plt.cm.spectral(labels[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    plt.xticks([])
    plt.yticks([])
    if title is not None:
        plt.title(title, size=17)
    plt.axis('off')
    plt.tight_layout()
```

**Excerise 1:** Apply PCA on the data

```python
### Your code here
from sklearn import decomposition
import numpy as np
pca = decomposition.PCA(n_components=2)
#fit_transform fits the data and reduces the dimensionality too. 
X_pca = pca.fit_transform(X)

plot_clustering(X_pca, X, y, title='PCA')
```

```python
X_pca[:10].shape
```

**Excerise 2:** Apply ISOMAP on the data

```python
### Your code here
from sklearn import manifold

iso = manifold.Isomap(n_neighbors=6, n_components=2)
iso.fit(X)
manifold_data = iso.transform(X)

plot_clustering(manifold_data, X, y, title='ISOMAP')
```

**Excerise 3:** Apply LLE on the data

```python
### Your code here

lle_data, _ = manifold.locally_linear_embedding(X, n_neighbors=12,
                                             n_components=2)
plot_clustering(lle_data, X, y, title='LLE')
```


**Excerise 4:** Apply t-SNE on the data

```python
### Your code here

tsne_data = manifold.TSNE(n_components=2).fit_transform(X)

plot_clustering(tsne_data, X, y, title='t-SNE')
```
